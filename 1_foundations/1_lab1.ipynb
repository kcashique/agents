```python
# Check the Gemini key (Ollama doesn't need a key).

import os
gemini_api_key = os.getenv('GEMINI_API_KEY')

if gemini_api_key:
    print(f"GEMINI API Key exists and begins {gemini_api_key[:8]}")
else:
    print("GEMINI API Key not set - please head to the troubleshooting guide in the setup folder")
    
```

```python
# And now - the all important import statement
# If you get an import error - head over to troubleshooting in the Setup folder
# This uses the new Gemini Python SDK

from google import genai
```

```python
# Configure Gemini with your API key
# If you're not sure what this means, check guide 6 for Python foundations

import random
import time

client = genai.Client(api_key=gemini_api_key)
model_nano = "gemini-2.5-flash"
model_mini = "gemini-2.5-flash"


def generate_with_retry(model, contents, max_retries=6):
    for attempt in range(1, max_retries + 1):
        try:
            return client.models.generate_content(model=model, contents=contents)
        except Exception as e:
            message = str(e)
            retryable = ("503 UNAVAILABLE" in message) or ("429 RESOURCE_EXHAUSTED" in message)
            if (not retryable) or attempt == max_retries:
                raise
            wait_seconds = min(60, (2 ** attempt) + random.random())
            print(f"Retrying after transient API error (attempt {attempt}/{max_retries}) in {wait_seconds:.1f}s...")
            time.sleep(wait_seconds)
```

```python
# Create a prompt for Gemini

prompt = "What is 2+2?"
```

```python
# And now call it! Any problems, head to the troubleshooting guide
# This uses Gemini 2.5 Flash

response = generate_with_retry(model=model_nano, contents=prompt)
print(response.text)
```

```python
# And now - let's ask for a question:

question_prompt = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
```

```python
# ask it - this also uses Gemini 2.5 Flash

response = generate_with_retry(model=model_mini, contents=question_prompt)
question = response.text

print(question)
```

```python
# use the generated question as the next prompt
answer_prompt = question
```

```python
# Ask it again

response = generate_with_retry(model=model_mini, contents=answer_prompt)
answer = response.text
print(answer)
```

```python
# Run 3 calls in sequence: Step 2 only runs if Step 1 succeeded, and Step 3 only runs if Step 2 succeeded.

import random
import time

# Guard rails for running this cell directly
if "client" not in globals() or "model_mini" not in globals():
    raise RuntimeError("Please run the setup cells first (key loading + Gemini client/model setup).")

if "generate_with_retry" not in globals():
    def generate_with_retry(model, contents, max_retries=6):
        for attempt in range(1, max_retries + 1):
            try:
                return client.models.generate_content(model=model, contents=contents)
            except Exception as e:
                message = str(e)
                retryable = ("503 UNAVAILABLE" in message) or ("429 RESOURCE_EXHAUSTED" in message)
                if (not retryable) or attempt == max_retries:
                    raise
                wait_seconds = min(60, (2 ** attempt) + random.random())
                print(f"Retrying after transient API error (attempt {attempt}/{max_retries}) in {wait_seconds:.1f}s...")
                time.sleep(wait_seconds)

business_idea = None
pain_point = None
solution = None

# Step 1 prompt
prompt_step_1 = """
You are an AI business strategist.

I am building a business in the Ready-to-Eat (RTE) food products industry.

Step 1:
Identify ONE specific business area within this industry that is worth exploring for an Agentic AI opportunity.

Requirements:
- Be specific (not generic like "marketing")
- Focus on real operations (supply chain, production, logistics, retail, etc.)
- Choose something where automation and decision-making matter

Output:
- Business Area Name
- Short Description (2-3 lines)
"""

try:
    print("Running Step 1...")
    response_1 = generate_with_retry(model=model_mini, contents=prompt_step_1)
    business_idea = (response_1.text or "").strip()

    if not business_idea:
        raise ValueError("Step 1 returned an empty response.")

    print("Step 1 complete.")

    # Step 2 prompt (only after Step 1 success)
    prompt_step_2 = f"""
You are an expert in operations and problem analysis.

Business Area:
{business_idea}

Step 2:
Identify ONE major pain-point or inefficiency in this business area that is suitable for an Agentic AI solution.

Requirements:
- Must be a real-world operational problem
- Should involve complexity, delays, or decision-making challenges
- Prefer problems involving cost, waste, or scalability

Output:
- Pain Point Title
- Detailed Explanation
- Why this is a serious business problem
"""

    print("Running Step 2...")
    response_2 = generate_with_retry(model=model_mini, contents=prompt_step_2)
    pain_point = (response_2.text or "").strip()

    if not pain_point:
        raise ValueError("Step 2 returned an empty response.")

    print("Step 2 complete.")

    # Step 3 prompt (only after Step 2 success)
    prompt_step_3 = f"""
You are an AI systems architect specializing in Agentic AI solutions.

Business Area:
{business_idea}

Pain Point:
{pain_point}

Step 3:
Propose a practical Agentic AI solution.

Requirements:
- The system must take autonomous decisions (not just dashboards)
- Explain how the AI agent behaves and acts
- Show step-by-step workflow

Output:
1. Solution Name
2. How the Agentic AI works (step-by-step)
3. Key decisions the AI makes autonomously
4. Data required
5. Expected business impact
6. Simple MVP idea
"""

    print("Running Step 3...")
    response_3 = generate_with_retry(model=model_mini, contents=prompt_step_3)
    solution = (response_3.text or "").strip()

    if not solution:
        raise ValueError("Step 3 returned an empty response.")

    print("Step 3 complete.")

except Exception as e:
    print(f"Workflow stopped: {e}")

# Show any successful outputs
if business_idea:
    print(f"\nBusiness Area:\n{business_idea}")

if pain_point:
    print(f"\nPain Point:\n{pain_point}")

if solution:
    print(f"\nAgentic AI Solution:\n{solution}")
```
