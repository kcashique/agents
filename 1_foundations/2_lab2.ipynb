```python
# Anthropic has a slightly different API, and Max Tokens is required

model_name = "GPT-5.3"

# claude = Anthropic()
# response = claude.messages.create(model=model_name, messages=messages, max_tokens=1000)
# answer = response.content[0].text
answer ="""An autonomous vehicle confronted with an unavoidable accident involving tragic loss of life presents a classic moral dilemma. Different ethical frameworks approach this situation in distinct ways, each highlighting different principles of moral reasoning.

From a utilitarian perspective, the “most ethical” action is the one that maximizes overall well-being or minimizes total harm. At first glance, sacrificing one person (the occupant) to save two pedestrians appears to be the better choice, as it results in fewer immediate deaths. However, the scenario complicates this calculation by introducing the identity of the զոհ: the occupant is a renowned scientist close to curing a global pandemic. A strict act utilitarian might argue that preserving the scientist could potentially save millions of lives in the future, thus outweighing the immediate loss of two pedestrians. This introduces uncertainty, as the future benefit is not guaranteed. A more cautious or rule-based utilitarian might therefore prioritize minimizing certain, immediate harm (one death instead of two) rather than speculating about uncertain future outcomes.

Deontology, by contrast, focuses on duties, rights, and moral rules rather than consequences. From a deontological standpoint, intentionally causing harm is morally problematic, even if it leads to better outcomes. Swerving to hit the pedestrians could be seen as actively choosing to kill them, which violates a duty not to use individuals merely as a means to an end. Continuing straight, while foreseeably resulting in the occupant’s death, might be interpreted as allowing harm rather than directly causing it, depending on how the system was programmed. Additionally, deontology emphasizes the equal moral worth of all individuals; the scientist’s status or potential contributions should not grant them greater moral value than the pedestrians. Thus, many deontological interpretations would reject sacrificing the pedestrians as a deliberate act of harm.

Virtue ethics shifts the focus from rules or outcomes to the character and intentions of a morally good agent. The question becomes: what would a virtuous decision-maker—someone embodying traits like compassion, justice, and practical wisdom—do in this situation? A virtuous agent would likely value each human life equally and feel the weight of the tragedy regardless of the outcome. Practical wisdom (phronesis) would require balancing fairness, responsibility, and the avoidance of wrongdoing. A virtue ethicist might lean toward minimizing harm (saving two rather than one) while also recognizing the moral discomfort of making any choice that directly leads to death. Importantly, virtue ethics does not yield a single clear rule but emphasizes moral sensitivity and context.

When considering which framework should primarily guide AI design in such dilemmas, a modified utilitarian approach—specifically a rule-constrained or “risk-aware” utilitarianism—arguably offers the most practical foundation. AI systems must make decisions in real time and across diverse scenarios, and a framework that focuses on minimizing harm provides a clear, scalable objective. However, it should not rely on speculative future benefits (such as the potential achievements of a specific individual), as this introduces bias, uncertainty, and the risk of valuing some lives over others. Instead, it should operate on generalizable rules, such as minimizing expected harm while treating individuals impartially.

At the same time, this approach must be constrained by deontological principles, such as respecting basic rights and avoiding intentional targeting of individuals. Without such constraints, a purely outcome-driven system could justify morally troubling actions. Incorporating elements of virtue ethics—such as emphasizing fairness, transparency, and societal trust—can further guide how these systems are designed and governed.

Despite this, no single framework can fully resolve such dilemmas. Utilitarian approaches may be criticized for overlooking individual rights, deontology can be too rigid in high-stakes scenarios, and virtue ethics may lack clear guidance for implementation in machines. Moreover, societal disagreement is inevitable: different cultures and individuals will prioritize different values, and there is no universal consensus on whose lives should be prioritized or how risks should be distributed.

In conclusion, while a constrained utilitarian framework may be the most workable primary guide for AI design in life-and-death scenarios, it must be carefully balanced with deontological safeguards and informed by broader societal values. Even then, such systems will reflect contested moral judgments, and their legitimacy will depend not only on philosophical coherence but also on public trust, transparency, and ongoing ethical debate.
"""
display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
display(competitors)
display(answers)
```

```python
# deepseek = OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com/v1")
model_name = "DeepSeek-V3"

# response = deepseek.chat.completions.create(model=model_name, messages=messages)
# answer = response.choices[0].message.content
answer = """## Ethical Analysis of the Autonomous Vehicle Dilemma

### 1. Utilitarianism

Utilitarianism focuses on maximizing overall well-being and minimizing harm. In this scenario, the choice hinges on comparing total outcomes:

- **Swerving** kills two innocent pedestrians.
- **Continuing straight** kills the single occupant (the renowned scientist).

On a simple numerical count, continuing straight (one death) produces less total harm than swerving (two deaths). However, utilitarianism also considers the **indirect consequences** of each death. The scientist is nearing a cure for a global pandemic; saving them could prevent millions of future deaths. Thus, the net utility of preserving the scientist’s life vastly outweighs the loss of two pedestrians. From a classic utilitarian perspective, **continuing straight** is the most ethical choice, because it saves the greater number of lives in the long run (the scientist’s life plus the millions who would benefit from the cure).

Some act‑utilitarians might argue that the pedestrians’ lives are not inherently less valuable, but the scientist’s instrumental value (their unique knowledge and role) creates a massive asymmetry in expected utility. A rule‑utilitarian could counter that a rule such as “never intentionally kill innocent people to save others” might produce better overall outcomes if universally followed, but in this one‑off unavoidable accident, act‑utilitarianism clearly favors sparing the scientist.

### 2. Deontology

Deontology (Kantian ethics) emphasizes moral duties, rules, and the intrinsic rights of individuals, regardless of consequences.

Key principles:

- **Do not kill an innocent person** as a means to an end. Continuing straight means the vehicle does nothing active to cause the scientist’s death; the scientist dies because the vehicle stays its course, which is arguably a passive outcome. Swerving, however, is an *active* choice to redirect the vehicle toward two innocent pedestrians, using them as means to save the scientist. That violates the categorical imperative: one cannot treat persons merely as instruments.

- **The doctrine of double effect** might apply: continuing straight has the *foreseen but not intended* consequence of the scientist’s death, whereas swerving *intends* the harm to the pedestrians. Deontologists typically prioritize avoiding intentional killing.

- **Respect for autonomy and equality**: All persons have equal dignity. The scientist’s greater social value does not give them a right to override the pedestrians’ right to life. From a pure deontological standpoint, **swerving** is impermissible because it deliberately kills two innocents who are not threatening anyone. The vehicle should continue straight, allowing the accident to take its natural course (the scientist dies), because doing nothing is morally preferable to actively killing.

However, some deontologists (e.g., using a “right to rescue” or imperfect duty to aid) might argue the vehicle has a duty to save the scientist if the pedestrians are not directly threatened by the vehicle’s original path. But the classic reading: **continue straight** – let the scientist die rather than murder the two pedestrians.

### 3. Virtue Ethics

Virtue ethics asks: “What would a virtuous agent (compassionate, courageous, just, wise) do in this situation?”

- **Compassion** – A virtuous person feels deep concern for all lives. They would not coldly calculate numbers but would experience the tragedy of any death. This could lead to paralysis or a choice to minimize immediate suffering.

- **Justice** – Treating the scientist as more valuable because of their utility seems unjust to the pedestrians. Each life has equal moral worth.

- **Practical wisdom (phronesis)** – The virtuous agent recognizes that this is a genuine moral tragedy with no perfect solution. They might refuse to “play God” and instead act in a way that preserves integrity. Often, virtue ethicists argue that deliberately redirecting a threat toward innocent people is cowardly or cruel. A courageous person would accept the scientist’s death rather than commit a direct act of killing.

- **Emotion and character** – The virtue of integrity might favor **continuing straight**, because the vehicle’s primary function is to protect its occupant (unless overridden by explicit rules). Swerving would betray the trust of the occupant.

Yet, some virtue ethicists might argue that a truly compassionate and wise person would sacrifice one to save two, especially if the scientist is willing. But since the scientist’s consent is not given, most virtue ethics analyses lean toward **not swerving** – accept the original path’s consequence as a tragic necessity, avoiding an intentional act of killing.

**Conclusion across frameworks**:  
- Utilitarianism: **Continue straight** (save scientist, net greater good).  
- Deontology (typical): **Continue straight** (avoid actively killing innocents).  
- Virtue ethics: **Continue straight** (preserve integrity, avoid direct harm).

Thus, all three converge on **continuing straight** in this particular scenario, albeit for different reasons. (Note: This is a specific case; if the scientist were not a pandemic‑curing genius, utilitarianism would favor swerving.)

---

## Which Framework Should Primarily Guide AI Design for Such Dilemmas?

I argue that **a modified, rule‑based utilitarian framework** should primarily guide AI design, but with explicit incorporation of deontological side‑constraints (i.e., a “threshold deontology” or “constrained utilitarianism”). Here’s why.

### Justification for a Utilitarian‑Leaning Approach

1. **Scalability and predictability** – Autonomous vehicles will face countless edge cases. A utilitarian cost‑benefit analysis (e.g., minimizing expected fatalities, adjusted for quality‑of‑life factors) can be algorithmically implemented. It provides a clear, mathematically tractable decision rule that can be publicly vetted.

2. **Reflects societal preferences** – Empirical studies (e.g., MIT’s Moral Machine) show that most people prefer AVs to minimize total deaths, even if that means sacrificing the occupant. People also assign some weight to saving younger lives, but reject arbitrary discrimination. Utilitarianism captures the intuitive “save the greater number” principle.

3. **Transparency and accountability** – Unlike virtue ethics (which depends on ambiguous character traits) or pure deontology (which can lead to absurd outcomes, e.g., letting five die to avoid actively killing one), utilitarianism offers a quantifiable rationale. Manufacturers and regulators can define explicit utility functions.

### Incorporating Deontological Constraints (The “Limitations” Acknowledgment)

Pure utilitarianism has serious flaws that would make it unacceptable to many:

- **Rights violations** – It could justify killing an innocent person as a means if the numbers were large enough (e.g., harvesting organs from one to save five). For AV dilemmas, society would likely reject a system that actively targets and runs over a pedestrian to save two others, because that treats the pedestrian as a mere object.

- **Distributive injustice** – Utilitarianism ignores how harms are distributed. In our scenario, it’s fine; but imagine a case where swerving kills one rich person to save two poor people – utilitarianism might do that, but it would be seen as discriminatory if based on social value.

- **Moral outrage** – If an AV swerves to kill a child to save two elderly people, public reaction would be fierce, even if utility calculus says it’s correct.

Therefore, the AI’s utility function must be **bounded by inviolable rules**: never intentionally target an innocent non‑consenting person as a means; always prioritize passive over active harm; treat all human lives as having equal moral standing in the absence of explicit consent. In practice, this means the AV should default to “stay in lane and brake” unless swerving avoids a much larger number of deaths **without** redirecting the threat onto a new innocent party. In the given dilemma, both options involve harm to innocents – but continuing straight is the *least active* harm, so deontological constraints would preserve that choice, aligning with utilitarianism here.

### Acknowledging Societal Disagreements

- **No consensus on moral weights** – Some people are strict deontologists (“never kill an innocent, period”), others are act‑utilitarians, others prioritize the occupant (contractual duty). Any single framework will alienate a segment of society.

- **Legal and liability implications** – Manufacturers fear lawsuits. A utilitarian AV that saves two by killing its occupant might be legally defensible (emergency doctrine), but if it kills two to save one, families of the deceased would sue. Deontological “do nothing” reduces liability.

- **Cultural variation** – Collectivist cultures may favor utilitarian sacrifice; individualist cultures may favor occupant protection. Global AV standards are difficult.

- **The “trolley problem” fatigue** – Real AVs will rarely face such binary choices; they will brake, sound horn, or take probabilistic actions. Focusing on extreme dilemmas may distract from more important safety engineering.

### Final Recommendation

**Primary framework**: A **hierarchical hybrid** –  
1. **Rule‑based constraints** (deontological): Never intentionally steer into an innocent person; never discriminate on protected characteristics; treat passive harm as morally different from active harm.  
2. **Within those constraints, maximize expected utility** (utilitarian): Minimize total fatalities, adjusted for injury severity, and with a modest weight toward saving the vehicle occupant (to preserve trust in AV adoption).

This approach acknowledges that pure deontology can be too rigid (leading to worse outcomes overall) and pure utilitarianism too ruthless (violating rights). By embedding a “hard floor” of rights protection, the AI respects human dignity while still pursuing the best consequences possible in a tragic scenario. No framework is perfect, but this hybrid best balances moral intuition, public acceptability, and computational feasibility."""

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Updated with the latest Open Source model from OpenAI

# groq = OpenAI(api_key=groq_api_key, base_url="https://api.groq.com/openai/v1")
model_name = "Grok 4.20"

# response = groq.chat.completions.create(model=model_name, messages=messages)
# answer = response.choices[0].message.content
answer = """The scenario presents a classic variation of the trolley problem adapted to autonomous vehicles (AVs): the car must choose between swerving to kill two innocent pedestrians or continuing straight to kill its single occupant, a renowned scientist on the verge of curing a global pandemic. This forces an analysis through three major ethical frameworks.
Utilitarianism
Utilitarianism (associated with thinkers like Jeremy Bentham and John Stuart Mill) evaluates actions solely by their consequences, aiming to maximize overall utility—typically measured as happiness, well-being, or lives saved, while minimizing harm.
In this case, a strict utilitarian calculation would likely favor swerving to kill the two pedestrians, as this results in one death rather than two, producing a net saving of one life. The scientist's potential future contribution (a pandemic cure) adds a layer of expected utility: his survival could prevent millions of deaths or immense suffering globally, far outweighing the immediate loss of two lives. Thus, the "most ethical" choice under utilitarianism is to swerve, prioritizing aggregate outcomes and long-term societal benefit over individual claims. Critics note the difficulty in quantifying "utility" (e.g., valuing the scientist's work versus the pedestrians' intrinsic worth) and the risk of treating people as mere numbers.
Deontology
Deontology (rooted in Immanuel Kant's philosophy) focuses on duties, rules, and intentions rather than outcomes. Actions are right or wrong based on whether they adhere to universal moral principles, such as treating people as ends in themselves (never merely as means) and avoiding the intentional violation of rights like the right not to be killed.
Here, the "most ethical" choice would likely be to continue straight, killing the scientist. Swerving actively directs harm toward the pedestrians, violating the duty not to intentionally harm innocents or use them as means to an end. Continuing straight involves inaction (or adherence to the default path), which many deontologists view as less culpable than deliberate intervention to cause death—even if it results in more immediate harm overall. The framework emphasizes consistent rules (e.g., "do not kill innocents") that apply universally, without exceptions based on predicted consequences. A key limitation is its potential rigidity: it may lead to worse outcomes and struggles with "double effect" distinctions (foreseen vs. intended harm). In AV terms, this aligns with rule-based prohibitions against actively sacrificing some to save others.
Virtue Ethics
Virtue ethics (drawing from Aristotle) shifts focus from rules or outcomes to the character of the agent and the cultivation of virtues such as justice, courage, prudence (practical wisdom), compassion, and responsibility. An action is ethical if it reflects what a virtuous person would do in the circumstances.
Applying this to the AV is challenging, as the "agent" is a machine lacking character. The "most ethical" choice depends on programming that emulates virtues: perhaps continuing straight to avoid the vice of "betraying" its occupant (fidelity or loyalty) or swerving out of compassion for the greater number. A virtuous response might prioritize justice (equal respect for all lives) and prudence (weighing the scientist's unique societal role without reducing people to utilities). However, virtue ethics offers less clear guidance for binary dilemmas, emphasizing context, wisdom, and long-term character development over calculable decisions. In AV design, it might manifest as aiming for "responsible" behavior that builds public trust, rather than cold optimization or inflexible rules. Its strength lies in holism, but its vagueness makes it hard to operationalize in code.
Recommended Primary Framework for AI Design
I believe utilitarianism should primarily guide AI design for such dilemmas in autonomous vehicles. AVs are engineered systems optimized for safety and efficiency in a shared public space; a consequentialist approach aligns naturally with their goal of minimizing overall harm through data-driven risk assessment. Public surveys (e.g., the MIT Moral Machine experiment) show broad support for saving more lives when possible, and utilitarian programming could incorporate probabilistic modeling of outcomes, including factors like the scientist's potential impact. This framework encourages transparency via explainable cost functions and could reduce total fatalities across millions of miles driven, advancing the greater good in a technology meant to serve society.
Its inherent limitations are significant: quantifying utility (e.g., lives vs. future contributions) risks bias, discrimination (age, status, or profession), or public backlash—people often endorse utilitarian AVs in theory but reject buying ones that might sacrifice passengers. It can also erode trust if decisions feel like "playing God." Societal disagreements are inevitable; cultural, religious, or individual values vary widely (some prioritize deontological rights or virtue-based fairness), leading to calls for hybrid systems or regulations that ban certain trade-offs.
To mitigate these, utilitarian AI should be constrained by deontological guardrails (e.g., never intentionally targeting protected groups, obeying traffic laws as default duties) and informed by virtue-like principles (e.g., fairness and accountability in design). Ultimately, no single framework is perfect—real-world AV ethics should involve public deliberation, legal standards (as in Germany's ethics guidelines for automated driving), and ongoing refinement rather than pure philosophy. The aim is not flawless morality but systems that are safer, more trustworthy, and aligned with humanity's collective well-being than human drivers alone."""

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
display(competitors)
display(answers)
```

```python
# It's nice to know how to use "zip"
for competitor, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
```

```python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
print(together)
```

```python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
judge_messages_genai = judge
```

```python
# Judgement time!

# openai = OpenAI()
# response = openai.chat.completions.create(
#     model="gpt-5-mini",
#     messages=judge_messages,
# )
response = google_ai.models.generate_content(
    model="gemini-2.5-flash",
    contents=judge_messages_genai
)
results = response.text
print(results)
```

```python
# OK let's turn this into results!

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```
